```python
import sqlalchemy
import pandas as pd
```

```python
db=sqlalchemy.create_engine('mysql+pymysql://root:pass@localhost:3306/mymovies')
```

```python
query="SELECT * FROM movies"
df=pd.read_sql_query(query,db)
df
```

## Pandas equivalent SQL queries

#### SELECT

```python
#SQL statement
'''SELECT * FROM movies;'''

#Pandas equivalent
df
```

```python
#SQL statement
'''SELECT name, year FROM movies;'''

#Pandas equivalent
df[['name','year']]
```

### SELECT DISTINCT (Finding distinct values of a column)

```python
#SQL statement
'''SELECT DISTINCT year FROM movies;'''

#Pandas equivalent
df['year'].unique()
```

### SELECT WHERE

```python
#SQL statement
'''SELECT * FROM movies WHERE rankscore>9;'''

#Pandas equivalent
df[df["rankscore"]>9]
```

### SELECT... WHERE ...AND

```python
#SQL statement
'''SELECT * FROM movies WHERE rankscore>9 AND year>2000;'''

#Pandas equivalent
df[(df['rankscore']>9) & (df['year']>2000)]
```

### SELECT... WHERE.... OR

```python
#SQL statement
'''SELECT * FROM movies WHERE rankscore>9 OR year>2000;'''

#Pandas equivalent
df[(df['rankscore']>9) | (df['year']>2000)]
```

### SELECT... WHERE...NOT

```python
#SQL statement
'''SELECT * FROM movies WHERE  NOT year<=2000;'''

#Pandas equivalent
df[df['year']>2000]
```

### IS NULL

```python
#SQL statement
'''SELECT * FROM movies WHERE rankscore IS NULL;'''

#Pandas equivalent
df[df['rankscore'].isnull()]
```

### IS NOT NULL

```python
#SQL statement
'''SELECT * FROM movies WHERE rankscore IS NOT NULL;'''

#Pandas equivalent
df[~df['rankscore'].isnull()]     # ~ --> symbol for NOT
```

### IN

```python
#SQL statement
'''SELECT * FROM movies WHERE year IN(2000,2002,2004);'''

#Pandas equivalent
df[df['year'].isin([2000,2002,2004])]
```

###  NOT IN

```python
#SQL statement
'''SELECT * FROM movies WHERE year NOT IN (2000,2002,2004);'''

#Pandas equivalent
df[~df['year'].isin([2000,2002,2004])]
```

### BETWEEN

```python
#SQL statement
'''SELECT * FROM movies WHERE year BETWEEN 1999 AND 2003;'''

#Pandas equivalent
df[(df['year']>=1999) & (df['year']<=2003)]
```

### AGGREGATE functions (MIN,MAX,COUNT,SUM,AVERAGE)

```python
#SQL statement
'''SELECT MIN(year) FROM movies;'''

#Pandas equivalent
df['year'].min()
```

```python
#SQL statement
'''SELECT MAX(year) FROM movies;'''

#Pandas equivalent
df['year'].max()
```

```python
#SQL statement
'''SELECT COUNT(year) FROM movies;'''

#Pandas equivalent
len(df['year'])
```

```python
#SQL statement
'''SELECT SUM(rankscore) FROM movies;'''

#Pandas equivalent
df['rankscore'].sum()
```

```python
#SQL statement
'''SELECT AVG(rankscore) FROM movies;'''

#Pandas equivalent
df['rankscore'].mean()
```

```python
#SQL statement
'''SELECT mean(rankscore), max(rankscore), min(rankscore), FROM movies;'''

#Pandas equivalent
df.agg({'rankscore':['mean','max','min']})
```

### ORDER BY(Ascending order)

```python
#SQL statement
'''SELECT *  FROM movies ORDER BY year;'''

#Pandas equivalent
df.sort_values(by='year')
```

### ORDER BY(Descending order)

```python
#SQL statement
'''SELECT *  FROM movies ORDER BY year DESC;'''

#Pandas equivalent
df.sort_values(by='year',ascending=False)
```

```python
#SQL statement
'''SELECT name,year FROM movies ORDER BY year DESC;'''

#Pandas equivalent
df[['name','year']].sort_values(by='year',ascending=False)
```

### LIMIT(SELECT with limited number of records)

```python
#SQL statement
'''SELECT * FROM movies LIMIT 20;'''

#Pandas equivalent
df.head(20)
```

```python
#SQL statement
'''SELECT name,rankscore FROM movies LIMIT 20;'''

#Pandas equivalent
df[['name','rankscore']].head(20)
```

```python
#SQL statement
'''SELECT * FROM movies WHERE rankscore>9 ORDER BY rankscore DESC LIMIT 20;'''

#Pandas equivalent
df[df['rankscore']>9].sort_values(by='rankscore',ascending=False).head(20)
```
